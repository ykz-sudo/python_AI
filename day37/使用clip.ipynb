```python
import torch

# 创建一个示例张量
x = torch.tensor([1.0, 3.5, 2.2, 7.7, 0.8])

# 取最大的3个元素及其索引
values, indices = torch.topk(x, k=3)

print("最大3个元素的值:", values)
print("最大3个元素的索引:", indices)
```

```python
import torch

# 创建一个2维张量
t = torch.tensor([[10, 20, 30], [40, 50, 60]])

# index张量指定了要采集的列索引
index = torch.tensor([[2, 1],     # 从第0行取第2列和第1列
                      [0, 2]])    # 从第1行取第0列和第2列

# 使用t.gather方法
result = t.gather(dim=1, index=index)

print("原始张量:\n", t)
print("采集索引:\n", index)
print("gather结果:\n", result)
```
