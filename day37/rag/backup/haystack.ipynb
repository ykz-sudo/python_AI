```python
import os
from haystack import Pipeline, PredefinedPipeline
import urllib.request

# 设置 OpenAI API 的密钥，用于后续调用
api_key = os.getenv("OPENAI_API_KEY")
if not api_key:
    raise ValueError("Please set OPENAI_API_KEY in your environment.")

# 下载并保存为 "davinci.txt"，这个文件包含了你要处理的文本内容
urllib.request.urlretrieve("https://archive.org/stream/leonardodavinci00brocrich/leonardodavinci00brocrich_djvu.txt",
                           "davinci.txt")

# 下载文件并保存为 "davinci.txt"，这个文件包含了将要处理的文本内容
urllib.request.urlretrieve("https://archive.org/stream/leonardodavinci00brocrich/leonardodavinci00brocrich_djvu.txt",
                           "davinci.txt")  

# 使用 Haystack 中的预定义索引管道来构建索引
# Pipeline.from_template 创建一个基于模板的管道实例
indexing_pipeline = Pipeline.from_template(PredefinedPipeline.INDEXING)

# 运行索引管道，将 "davinci.txt" 文件作为数据源导入管道
# 这样可以在后续查询时检索这部分内容
indexing_pipeline.run(data={"sources": ["davinci.txt"]})

# 创建一个 RAG（Retrieval-Augmented Generation）管道实例
# RAG管道用于检索增强生成，结合检索和生成能力
rag_pipeline = Pipeline.from_template(PredefinedPipeline.RAG)

# 定义查询问题，即要检索的内容
query = "How old was he when he died?"

# 运行 RAG 管道，生成回答。设置 `data` 参数，其中包含 "prompt_builder" 和 "text_embedder" 配置：
# "prompt_builder" 指定问题内容，"text_embedder" 负责生成所需的文本表示。
result = rag_pipeline.run(data={"prompt_builder": {"query": query}, "text_embedder": {"text": query}})

# 输出 RAG 管道的第一个生成回答
print(result["llm"]["replies"][0])
```
