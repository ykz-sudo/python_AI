```python
s1=pd.Series(['A','B','C','D'],index=['001','002','003','004'])
print(s1)
print(s1[0])  #未来版本中，s1[0]将被视为标签，而不是位置索引
print(s1['002'])

#推荐
print(s1.iloc[2])
print(s1.loc['002'])
```

```python
df = pd.DataFrame(np.arange(12).reshape(3,4),index=['s1','s2','s3'],columns=['c1','c2','c3','c4'])
print("原始数据:")
print(df)

# 1. 通过列索引获取值
print("通过列索引获取值:")
print(df[['c2']])
print(type(df[['c2']]))
print(df['c2'])
print(type(df['c2']))

# 2. 通过多个列索引获取多个列的值
print("通过多个列索引获取多个列的值")
print(df[['c2','c4']])

# 3. 通过行索引获取行数据
print("通过行索引获取行数据:")
print(df.loc['s2'])
```

```python
# 4. 通过多个行索引获取多个行数据
print("通过多个行索引获取多个行数据:")
print(df.loc[['s1','s3']])

# 5. 通过位置索引获取一行的值
print("通过位置索引获取一行的值:")
print(df.iloc[1])

# 6. 通过位置索引获取多行数据
print("通过位置索引获取多行数据:")
print(df.iloc[[0,2]])

# 7. 获取具体位置的值
print("获取具体位置的值:")
print(df.iloc[1,2])         # 前面是行，后面是列


# 8. 通过位置索引切片获取数据
print("通过位置索引切片获取数据:")
print(df.iloc[0:2,1:3])     # 前面是行，后面是列，左闭右开

# 9. 通过名字索引切片获取数据
print("通过名字索引切片获取数据:")
print(df.loc['s1':'s3','c2':'c4'])  # 前面是行，后面是列，左闭右闭

# 10. 一次取多个不连续的行或者列
print(df.iloc[[1,2],[0,3]])  # 前面是行，后面是列
#改为loc实现上一行
print(df.loc[['s2','s3'],['c1','c4']])

```

## 4、完成apply和map的练习

```python
# 使用案例
# 创建一个学生表
data = {
    "name": ["Alice", "Bob", "Charlie", "David"],
    "age": [20, 22, 21, 20],
    "score": [85, 90, 78, 92],
    "gender": ["女", "男", "男", "男"]
}
df = pd.DataFrame(data, index=['s1', 's2', 's3', 's4'], columns=['name', 'age', 'score', 'gender'])
print(df)

# 比如需要把年龄数据乘以2
def mul_two(x):
    return x * 2
df['age'] = df['age'].apply(mul_two)
# 再次打印一下看看变化
print(df)

# 也可以直接使用匿名函数
df['age'] = df['age'].apply(lambda x: x * 2)
print(df)
```

```python
#处理DataFrame
# apply也可以处理DataFrame数据
# 和Series的区别在于，操作DataFrame的时候要指定轴

df = pd.DataFrame(np.arange(12).reshape(3,4), index=['s1', 's2', 's3'], columns=['c1', 'c2', 'c3', 'c4'])
print(df)

# 1. 统计每个列的数据个数
print(df.apply(lambda x: x.count(), axis=0))        # 默认轴是0，即列

# 2. 统计每个列的最大值
print(df.apply(lambda x: x.max(), axis=0))          # 默认轴是0，即列
print(type(df.apply(lambda x: x.max(), axis=0)))

# 3. 统计每个行的数据个数
print(df.apply(lambda x: x.count(), axis=1))        # 轴1, 是行
```

## 5、完成groupby与agg的联系

```python

# 创建一个数据表
data = {
    "name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank"],
    "age": [20, 22, 21, 20, 20, 30],
    "score": [85, 90, 88, 92, 85, 72],
    "gender": ["女", "男", "男", '男', '女', '男']
}
df = pd.DataFrame(data, index=['s1', 's2', 's3', 's4', 's5', 's6'], columns=['name', 'age', 'score', 'gender'])
print(df)

# 1. 单列统计
print("单列统计:")
print(df['age'].describe())
print('-'*20)

# 2. 多列统计
print("多列统计:")
print(df[['age', 'score']].describe())
print('-'*20)

# 3. 所有列进行统计(只会对，也只能对数值列进行统计)
print('所有列统计')
print(df.describe(include='all'))

# 4. 常用统计函数
print("常用统计函数:")
print(df['age'].mean())             # 平均数
print(df['age'].median())           # 中位数
print(df['age'].min())              # 最小值
print(df['age'].max())              # 最大值
print(df['age'].std())              # 标准差
print(df['age'].sum())              # 求和
```

```python
# 1. 单列多函数聚合
score_stats = df['score'].agg([
    "mean", "max", "min"
])
print(score_stats)
print(type(score_stats))

# 2. 多列分别聚合
multi_col_stats = df.agg({
    "age": ["mean", "count"],    # age列：均值、非空计数
    "score": ["median", "std"]   # score列：中位数、标准差
})
print(multi_col_stats)
```

```python
# 创建一个数据表
data = {
    "name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank"],
    "age": [20, 22, 21, 20, 20, 30],
    "score": [85, 90, 88, 92, 82, 72],
    "gender": ["女", "男", "男", '男', '女', '男']
}
df = pd.DataFrame(data, index=['s1', 's2', 's3', 's4', 's5', 's6'], columns=['name', 'age', 'score', 'gender'])
print(df)

# 1. 分组
group_by_gender = df.groupby('gender')
print(group_by_gender)
print(type(group_by_gender))
# 2. 分组之后统计
print("分组之后统计:")
print(group_by_gender.describe())


# 3. 分组之后求均值
print("分组之后求均值:")
mean_score = df.groupby('gender')['score'].mean().add_prefix('mean_')
print(mean_score)
print(type(mean_score))
```

```python
# 定义自定义极值函数
def gender_extremum(series):
    return f"最大值:{series.max()} 最小值:{series.min()}"

# 按性别分组，对score求极值（自定义函数）
gender_score_extremum = df.groupby('gender')['score'].agg(gender_extremum)
print(gender_score_extremum)
```

```python
# 4. 分组之后进行转换
df["score_mean"] = df.groupby("gender")["score"].transform('mean')
print(df[["name", "gender", "score", "score_mean"]])
df["score_minus_mean"] = df["score"]-df.groupby("gender")["score"].transform('mean')
print(df)
print(df.groupby("gender")["score"])

# 需求：按性别分组，计算每个学生分数在组内的降序排名(传入自定义函数)
# ascending=False 表示降序
df["score_rank"] = df.groupby("gender")["score"].transform(lambda group: group.rank(ascending=False))
print(df[["name", "gender", "score", "score_rank"]])
```

1)机器学习的训练过程，是指利用已有数据，通过算法不断调整模型参数，使模型学会输入（特征）与输出（标签）之间的映射关系，从而在新数据上做出较准确预测的过程。
一般包括：数据准备 → 模型选择 → 训练 → 评估 → 调优 → 测试。
2)特征是用来描述样本属性的输入变量，是模型进行学习和预测的依据。
3)标签是样本对应的真实输出结果，也是模型学习的目标。
4)训练集是用于训练模型的数据集合，模型通过训练集学习特征与标签之间的关系，并调整自身参数。
5)测试集是用于评估模型性能的数据集合，在模型训练完成后使用，用来检验模型在未见过数据上的泛化能力。
6)监督学习是指训练数据中每个样本都包含特征和对应标签，模型通过已知正确答案进行学习。
常见任务包括：分类和回归。
7)无监督学习是指训练数据没有标签，模型通过挖掘数据内部结构或分布规律进行学习。
常见任务包括：聚类、降维。
8)分类问题是指模型预测的输出是离散类别。
9)回归问题是指模型预测的输出是连续数值。
10)过拟合是指模型在训练集上表现很好，但在测试集或新数据上表现很差，说明模型过度学习了训练数据中的噪声和细节，泛化能力不足。
11)欠拟合是指模型过于简单，无法很好地学习训练数据中的规律，在训练集和测试集上表现都较差。
