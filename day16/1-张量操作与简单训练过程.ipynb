````python
import torch

x = torch.randn(2, 3, 4)
print("{获取张量形状}:", x.shape)
print("{获取张量大小}:", x.size())         # size() 方法和 shape 一样，返回张量的形状

# 1. 调整形状
x = x.reshape(2, 12)            # 调整为2行12列
print("{调整为2行12列}:",x.shape)
x = x.reshape(-1, 3)                # 不限制行数，列数调整为3列（会自动计算行数）
print("{自动计算行数，3列}:",x.shape)


# 2. 交换维度
x = torch.randn(2, 3, 4)
x = x.transpose(1, 2)
print("{交换第1和第2维度}:", x.shape)

# 3. 重新排列多个维度
# 第3维度变为第1维，第2维变为第2维，第1维变为第3维
x = x.permute(2,1,0)
print("{重新排列维度(2,1,0)}:", x.shape)

# 4. 在指定维度上增加一个维度， 会添加形状为1的维度
x = x.unsqueeze(dim = 2)
print("{在第2维增加维度}:", x.shape)

# 5. 在指定维度上删除一个维度， 会删除形状为1的维度（如果该维度不存在形状为1的维度，则返回原张量）
x = x.squeeze(dim = 2)
print("{删除第1维的维度}:", x.shape)

# 6. 调整形状，需要连续内存
# 经过升维和降维操作，张量的数据存储位置（内存）可能会改变，所以内存可能不连续
# 使用view方法，如果内存不连续，那么会报错
# x = x.view(3,-1)
# print(f'view后 {x.shape}')

# 6. 判断是否是连续内存
print("{判断内存是否连续}:", x.is_contiguous())

# 7. 转换为连续内存
x = x.contiguous()
print("{转换后内存是否连续}:", x.is_contiguous())

# 8. 调整形状，需要连续内存
x = x.view(3,-1)
print("{使用view调整形状}:", x.shape)
````

# view的数据共享问题

````python
# 9. 验证view操作是否共享内存
x = torch.randn(2, 3, 4)
print("{原始张量形状}:", x.shape)
print("{原始张量}:\n", x)

# 使用view调整形状
x_view = x.view(6, 4)
print("{view后张量形状}:", x_view.shape)

# 验证是否共享内存 - 修改原张量
x[0, 0, 0] = 999
print("{修改原张量后，原张量第一个元素}:", x[0, 0, 0])
print("{修改原张量后，view张量第一个元素}:", x_view[0, 0])

# 验证是否共享内存 - 修改view张量
x_view[1, 1] = 888
print("{修改view张量后，view张量[1,1]}:", x_view[1, 1])
print("{修改view张量后，原张量对应位置}:", x[0, 1, 1])

# 检查内存地址是否相同
print("{原张量id}:", id(x))
print("{view张量id}:", id(x_view))

print("{原张量数据指针}:", x.data_ptr())
print("{view张量数据指针}:", x_view.data_ptr())
print("{是否共享存储}:", x.storage().data_ptr() == x_view.storage().data_ptr())
````

# torch中导数（梯度）计算

````python
import torch

# 1. 准备数据: 参数与目标值
x = torch.tensor(5) #特征
y = torch.tensor(0.)

# 2. 初始化参数: 权重与偏置
w = torch.tensor(1., requires_grad=True, dtype=torch.float32)
b = torch.tensor(3., requires_grad=True, dtype=torch.float32)

# 3. 构建神经网络并输出值
z = w * x + b

# 4. 计算损失
loss =  torch.nn.MSELoss()
loss = loss(z, y)

# 5. 计算梯度
loss.backward()

# 6. 访问梯度
print(w.grad)
print(b.grad)
````

````python
# 线性回归案例

import torch
import torch.nn as nn
import torch.optim as optim
from  torch.utils.data import TensorDataset
from  torch.utils.data import DataLoader
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

# 1. 生成数据
# 100个样本，1个特征，噪声为10，bias为1.5, coef=True: 返回基础线性回归的系数
x,y,coef = make_regression(n_samples=100, n_features=1, noise=10, bias=1.5, coef=True, random_state=22)

# 2. 数据可视化
# plt.scatter(x,y)

# 3. 使用Pytorch模拟线性回归
# 3.1 构造数据集对象 Dataset
dataset = TensorDataset(torch.from_numpy(x), torch.from_numpy(y))

# 3.2 创建数据加载器 DataLoader
# shuffle=:样本数据是否进行乱序
# batch_size: 每次迭代的样本数量
dataLoader = DataLoader(dataset, batch_size=10, shuffle=True)

# 3.3 构造模型
model = nn.Linear(in_features=1, out_features=1)

# 3.4 构造损失函数, 优化函数
loss_fn = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# 3.5 训练模型
epochs = 100        # 迭代次数
losses = []         # 损失值列表
total_loss = 0.     # 损失值
train_sample = 0.   # 训练样本数

for epoch in range(epochs):
    for train_x, train_y in dataLoader:

        # 训练模型，预测结果（注意类型转换）
        y_pred = model(train_x.type(torch.float32))

        # 计算损失（注意类型转换）
        loss = loss_fn(y_pred, train_y.reshape(-1,1).type(torch.float32))

        # 反向传播，更新参数
        optimizer.zero_grad()       # 清空梯度
        loss.backward()             # 反向传播
        optimizer.step()            # 更新参数

        # print("epoch:{}, loss:{}".format(epoch, loss.item()))
        total_loss += loss.item()
        train_sample += len(train_y)

    # 单个epoch的平均训练损失
    losses.append(total_loss / train_sample)
# 4. 绘制损失变化曲线
plt.plot(range(epochs), losses)
plt.title("loss curve")
plt.show()

# 5. 绘制拟合曲线
plt.scatter(x,y)            # 数据散点图

x = torch.linspace(x.min(), x.max(), 100)
y_train = torch.tensor([v*model.weight + model.bias for v in x])        # 使用训练的权重和偏置
y_true = torch.tensor([v*coef + 1.5 for v in x])                        # 使用真实的权重和偏置

# 绘制曲线
plt.plot(x, y_train, label = 'train')
plt.plot(x, y_true, label = 'true')
plt.grid()                              # 绘制网格
plt.legend()                            # 绘制图例
plt.show()
````

## ========== PyTorch线性回归案例：函数方法详解 ==========

### 📚 本案例使用的所有重要函数方法汇总：

| 函数/方法 | 功能 | 语法示例 | 返回值类型 | 详细说明 |
|---------|------|---------|-----------|------|
| **`make_regression()`** | 生成回归数据 | `make_regression(n_samples=100, n_features=1, noise=10)` | `(X, y, coef)` | sklearn生成线性回归模拟数据 |
| **`torch.from_numpy()`** | numpy转张量 | `torch.from_numpy(array)` | `torch.Tensor` | 将numpy数组转换为PyTorch张量 |
| **`TensorDataset()`** | 创建数据集 | `TensorDataset(X_tensor, y_tensor)` | `Dataset对象` | 将特征和标签打包成数据集 |
| **`DataLoader()`** | 创建数据加载器 | `DataLoader(dataset, batch_size=10, shuffle=True)` | `DataLoader对象` | 批量加载数据，支持shuffle |
| **`nn.Linear()`** | 线性层 | `nn.Linear(in_features=1, out_features=1)` | `Linear对象` | 创建线性回归模型 y=wx+b |
| **`nn.MSELoss()`** | 均方误差损失 | `nn.MSELoss()` | `MSELoss对象` | 计算MSE损失：(1/n)Σ(y_pred-y_true)² |
| **`optim.SGD()`** | 随机梯度下降 | `optim.SGD(model.parameters(), lr=0.01)` | `SGD对象` | 优化器，lr是学习率 |
| **`.parameters()`** | 获取模型参数 | `model.parameters()` | `参数生成器` | 返回模型所有可训练参数 |
| **`model.forward()`** | 前向传播 | `model(x)` | `torch.Tensor` | 模型预测：y_pred = x @ w + b |
| **`.zero_grad()`** | 梯度清零 | `optimizer.zero_grad()` | `None` | 清空上一步的梯度，避免累积 |
| **`.backward()`** | 反向传播 | `loss.backward()` | `None` | 计算所有参数的梯度 |
| **`.step()`** | 参数更新 | `optimizer.step()` | `None` | 根据梯度更新模型参数 |
| **`.item()`** | 张量转数值 | `loss.item()` | `float` | 将标量张量转换为Python数值 |
| **`torch.linspace()`** | 等差数列 | `torch.linspace(start, end, steps)` | `torch.Tensor` | 创建等间距的数值序列 |
| **`.detach()`** | 分离计算图 | `tensor.detach()` | `torch.Tensor` | 从计算图中分离，避免梯度追踪 |
| **`.numpy()`** | 张量转numpy | `tensor.numpy()` | `numpy.ndarray` | 将PyTorch张量转换为numpy数组 |

### 🔧 关键概念详解：

#### 1. **张量（Tensor）与自动求导（Autograd）**
- **张量**：PyTorch的核心数据结构，多维数组，支持GPU加速
- **自动求导**：PyTorch自动计算梯度的机制，通过`.backward()`实现
- **计算图**：记录张量运算的图结构，用于反向传播

#### 2. **Dataset与DataLoader**
- **Dataset**：数据封装类，提供`__getitem__`和`__len__`方法
- **DataLoader**：数据加载器，支持批量加载、多进程、shuffle等
- **batch_size**：每个批次的大小，影响内存使用和训练稳定性
- **shuffle**：打乱数据顺序，避免模型记住样本顺序

#### 3. **模型参数（Parameters）**
- **权重（weight）**：线性回归的斜率，存储在`model.weight`
- **偏置（bias）**：线性回归的截距，存储在`model.bias`
- **requires_grad**：参数是否需要梯度，默认True

#### 4. **损失函数（Loss Function）**
- **MSELoss**：均方误差，回归问题的标准损失函数
- **公式**：`MSE = (1/n) * Σ(y_pred - y_true)²`
- **梯度**：`∂MSE/∂w = (2/n) * Σ(y_pred - y_true) * x`

#### 5. **优化器（Optimizer）**
- **SGD**：随机梯度下降，最基本的优化算法
- **学习率（lr）**：控制每次参数更新的步长
- **动量（momentum）**：加速收敛，减少震荡

#### 6. **训练循环（Training Loop）**
```python
# 标准训练步骤：
for epoch in range(epochs):          # 外层：遍历epochs
    for batch in dataLoader:         # 内层：遍历batches
        # 1. 前向传播
        y_pred = model(x)
        # 2. 计算损失
        loss = loss_fn(y_pred, y_true)
        # 3. 反向传播
        optimizer.zero_grad()         # 梯度清零
        loss.backward()               # 计算梯度
        optimizer.step()              # 更新参数
```

#### 7. **梯度下降原理**
- **目标**：最小化损失函数 `L(w, b)`
- **梯度**：函数的导数，表示最陡峭上升方向
- **更新规则**：
  - `w = w - lr * ∂L/∂w`
  - `b = b - lr * ∂L/∂b`

### 🎯 实际应用建议：

1. **数据预处理**：标准化特征可以加速收敛
2. **学习率调优**：太大震荡，太小收敛慢
3. **批大小选择**：太大内存不足，太小训练不稳定
4. **监控训练**：观察损失曲线判断训练状态
5. **早停机制**：当验证损失不再下降时停止训练

### 📊 可视化解读：

- **损失曲线**：理想情况下应该单调递减并趋于平缓
- **拟合曲线**：训练曲线应该接近真实曲线，但不会完全重合（因为有噪声）
- **散点分布**：数据应该大致呈线性分布

这个案例展示了PyTorch深度学习框架的核心工作流程：数据准备 → 模型构建 → 训练循环 → 结果评估。
