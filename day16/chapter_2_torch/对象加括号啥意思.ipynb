```python
class Adder:
    def __init__(self, initial_value=0):
        # 初始化一个初始值
        self.value = initial_value

    def __call__(self, x):
        # 当实例被调用时，将传入的值加到初始值上
        self.value += x
        return self.value

# 创建一个 Adder 的实例
adder = Adder(10)

# 调用实例，就像调用函数一样
print(adder(5))  # 输出: 15
print(adder(3))  # 输出: 18
print(adder(2))  # 输出: 20
```
