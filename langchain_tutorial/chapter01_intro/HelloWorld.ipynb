## 1、获取大模型

dotenv.load_dotenv() 只会加载当前工作目录下的 .env 文件，也就是你运行 Python 脚本时所在的目录。

如果当前目录找不到 .env，它会继续往 父目录 一层一层向上查找，直到找到第一个 .env 文件，或者到达根目录为止。

```python
#导入 dotenv 库的 load_dotenv 函数，用于加载环境变量文件（.env）中的配置
import dotenv
from langchain_openai import ChatOpenAI
import os

dotenv.load_dotenv()  #加载当前目录下的 .env 文件



os.environ['OPENAI_API_KEY'] = os.getenv("OPENAI_API_KEY1")
os.environ['OPENAI_BASE_URL'] = os.getenv("OPENAI_BASE_URL")

# 创建大模型实例
#这个模型不支持自定义temperature参数，只支持默认值1
llm = ChatOpenAI(model="gpt-5-nano", temperature=1)  # 显式设置temperature=1

# 直接提供问题，并调用llm
response = llm.invoke("什么是大模型？")
print(response)
```

## 2、 使用提示词模板



```python
from langchain_core.prompts import ChatPromptTemplate  # 导入ChatPromptTemplate类，用于创建聊天提示模板

# 需要注意的一点是，这里需要指明具体的role，在这里是system和用户
prompt = ChatPromptTemplate.from_messages([  # 创建一个聊天提示模板，从消息列表初始化
    ("system", "你是世界级的技术文档编写者"),  # 系统角色提示，定义AI的行为方式
    ("user", "{input}")  # {input}为变量，用户输入的占位符
])

# 我们可以把prompt和具体llm的调用和在一起。
chain = prompt | llm  # 使用管道操作符将提示模板和语言模型连接起来，形成一个处理链
message = chain.invoke({"input": "大模型中的LangChain是什么?"})  # 调用处理链，传入输入参数，获取响应
print(message)  # 打印模型的响应结果

# print(type(message))  # 注释掉的代码，用于打印message的类型
```

## 3、 使用输出解析器：JSONOutputParser

```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser,JsonOutputParser

# 初始化模型
llm = ChatOpenAI(model="gpt-4o-mini")

# 创建提示模板
prompt = ChatPromptTemplate.from_messages([
    ("system", "你是世界级的技术文档编写者。输出格式要求：{format_instructions}"),
    ("user", "{input}")
])

# 使用输出解析器
# output_parser = StrOutputParser()
output_parser = JsonOutputParser()

# 将其添加到上一个链中
# chain = prompt | llm
chain = prompt | llm | output_parser

# 调用它并提出同样的问题。答案是一个字符串，而不是ChatMessage
# chain.invoke({"input": "LangChain是什么?"})
chain.invoke({"input": "LangChain是什么? ","format_instructions":output_parser.get_format_instructions()})
```

## 4、使用向量数据库存储



```python
# 导入WebBaseLoader类，用于从网页加载文档
from langchain_community.document_loaders import WebBaseLoader
# 导入bs4库，用于解析HTML内容
import bs4
os.environ["USER_AGENT"] = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/118.0 Safari/537.36"
# 创建WebBaseLoader实例，指定网页URL和BeautifulSoup解析参数
loader = WebBaseLoader(
        web_path="https://www.gov.cn/xinwen/2020-06/01/content_5516649.htm",  # 指定要加载的政府网站URL
        bs_kwargs=dict(parse_only=bs4.SoupStrainer(id="UCAP-CONTENT"))  # 只解析id为UCAP-CONTENT的内容部分
    )
# 加载网页内容并转换为文档对象
docs = loader.load()
# print(docs)  # 注释掉的打印语句，用于调试时查看加载的文档内容

# 导入OpenAI的嵌入模型类
from langchain_openai import OpenAIEmbeddings

# 创建嵌入模型实例，使用OpenAI的text-embedding-ada-002模型
embeddings = OpenAIEmbeddings(model="text-embedding-ada-002")


# 导入FAISS向量存储类，用于存储和检索向量
from langchain_community.vectorstores import FAISS
# 导入文本分割器，用于将长文本分割成小块
from langchain_text_splitters import RecursiveCharacterTextSplitter

# 创建文本分割器实例，设置每块大小为500字符，相邻块重叠50字符
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
# 使用分割器将文档分割成小块
documents = text_splitter.split_documents(docs)
# 打印分割后的文档块数量
print(len(documents))
# 创建FAISS向量存储，将文档向量化并存储
# embeddings会将documents中的每个文本片段转换为向量，并将这些向量存储在FAISS向量数据库中（默认会存储在内存中）
vector = FAISS.from_documents(documents, embeddings)
```

## 5、使用RAG进行检索



```python
# 导入PromptTemplate类，用于创建提示词模板
from langchain_core.prompts import PromptTemplate

# 将向量存储转换为检索器对象，vector就是我们上面得到的
retriever = vector.as_retriever()
# 设置检索参数，k=3表示每次检索返回3个最相关的文档
retriever.search_kwargs = {"k": 3}
# 使用检索器查询"建设用地使用权是什么？"，获取相关文档
docs = retriever.invoke("建设用地使用权是什么？")

# 以下是注释掉的代码，用于遍历并打印检索到的文档
# for i,doc in enumerate(docs):
#     print(f"⭐第{i+1}条规定：")
#     print(doc)

# 6.定义提示词模版
prompt_template = """
你是一个问答机器人。
你的任务是根据下述给定的已知信息回答用户问题。
确保你的回复完全依据下述已知信息。不要编造答案。
如果下述已知信息不足以回答用户的问题，请直接回复"我无法回答您的问题"。

已知信息:
{info}

用户问：
{question}

请用中文回答用户问题。
"""
# 7.得到提示词模版对象
# 从字符串创建PromptTemplate对象
template = PromptTemplate.from_template(prompt_template)

# 8.得到提示词对象
# 将检索到的文档和问题填入模板，生成完整提示词
prompt = template.format(info=docs, question='建设用地使用权是什么？')

## 9. 调用LLM
# 使用语言模型处理提示词并获取响应
response = llm.invoke(prompt)
# 打印语言模型的回答内容
print(response.content)
```

## 6、 创建Agent的使用

```python
from langchain.tools.retriever import create_retriever_tool  # 导入创建检索工具的函数

# 创建一个基于检索器的工具
retriever_tool = create_retriever_tool(
    retriever,  # 使用之前定义的检索器
    "CivilCodeRetriever",  # 工具的名称
    "搜索有关中华人民共和国民法典的信息。关于中华人民共和国民法典的任何问题，您必须使用此工具!",  # 工具的描述
)

tools = [retriever_tool]  # 将检索工具添加到工具列表中

from langchain import hub  # 导入hub模块，用于获取预定义的提示模板
from langchain.agents import create_openai_functions_agent  # 导入创建OpenAI函数代理的函数
from langchain.agents import AgentExecutor  # 导入代理执行器类

# 从LangChain Hub获取预定义的代理提示模板 https://smith.langchain.com/hub
prompt = hub.pull("hwchase17/openai-functions-agent")  # 从hub拉取预定义的代理提示模板

agent = create_openai_functions_agent(llm, tools, prompt)  # 创建OpenAI函数代理，传入语言模型、工具和提示模板
agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True)  # 创建代理执行器，设置verbose=True以显示执行过程

# 调用代理执行器处理查询
agent_executor.invoke({"input": "建设用地使用权是什么"})  # 向代理提交关于建设用地使用权的查询
```
