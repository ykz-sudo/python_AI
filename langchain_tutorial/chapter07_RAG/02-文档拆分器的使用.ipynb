# 文档拆分器的使用

## 1、CharacterTextSplitter

````python
# 1.导入相关依赖
from langchain.text_splitter import CharacterTextSplitter

# 2.示例文本
text = """
LangChain 是一个用于开发由语言模型驱动的应用程序的框架的。它提供了一套工具和抽象，使开发者能够更容易地构建复杂的应用程序。
"""

# 3.定义字符分割器,CharacterTextSplitter是 按字符数分割
splitter = CharacterTextSplitter(
    chunk_size=51, # 每块大小
    chunk_overlap=5,# 块与块之间的重复字符数
    #length_function=len, # 计算字符数
    separator=""   # 设置为空字符串时，表示禁用分隔符优先
)

# 4.分割文本
# split_text()的参数是字符串，返回的是字符串列表
texts = splitter.split_text(text)

# 5.打印结果
for i, chunk in enumerate(texts):
    print(f"块 {i+1}:长度：{len(chunk)}")
    print(chunk)
    print("-" * 50)
````

# `keep_separator` 参数详解

## 一、基本功能

`keep_separator` 参数用于控制**分隔符是否保留在分割后的文本块中**。

- `keep_separator=True`（默认）：**保留**分隔符
- `keep_separator=False`：**删除**分隔符

## 二、对比示例

### 原始文本
```python
text = "人工智能是未来趋势。机器学习是AI的核心。深度学习发展迅速。神经网络应用广泛。"
```

### 情况 1：`keep_separator=True`（默认）

```python
from langchain_text_splitters import CharacterTextSplitter

text_splitter = CharacterTextSplitter(
    chunk_size=30,
    chunk_overlap=5,
    separator="。",
    keep_separator=True  # 保留句号
)

chunks = text_splitter.split_text(text)
print(chunks)
```

**输出：**
```python
[
    '人工智能是未来趋势。',
    '趋势。机器学习是AI的核心。',
    '核心。深度学习发展迅速。',
    '迅速。神经网络应用广泛。'
]
```

**特点：**
- ✅ 句号被**保留**在每个块中
- ✅ 保持了句子的完整性和语义边界
- ✅ 阅读时有明确的句子结束标记

---

### 情况 2：`keep_separator=False`

```python
text_splitter = CharacterTextSplitter(
    chunk_size=30,
    chunk_overlap=5,
    separator="。",
    keep_separator=False  # 删除句号
)

chunks = text_splitter.split_text(text)
print(chunks)
```

**输出：**
```python
[
    '人工智能是未来趋势',
    '趋势机器学习是AI的核心',  # 注意：没有句号，两个句子直接连接
    '核心深度学习发展迅速',
    '迅速神经网络应用广泛'
]
```

**特点：**
- ❌ 句号被**删除**
- ❌ 句子之间没有明确分隔，可能影响语义理解
- ⚠️ 在 overlap 区域可能造成歧义

---

## 三、详细对比表

| 特性 | `keep_separator=True` | `keep_separator=False` |
|------|----------------------|------------------------|
| **分隔符保留** | 保留在文本块中 | 完全删除 |
| **语义完整性** | 好（句子边界清晰） | 差（句子可能连在一起） |
| **字符计数** | 包含分隔符 | 不包含分隔符 |
| **适用场景** | 自然语言文本 | 结构化数据（如 CSV） |
| **可读性** | 高 | 低 |

---

## 四、实际应用场景

### 场景 1：销售对话文本（推荐 `True`）

**优势：**
- ✅ 保持对话的自然断句
- ✅ 便于后续检索和阅读

## 五、与 `chunk_overlap` 的交互

### 示例：重叠区域的差异

```python
text = "第一句话。第二句话。第三句话。第四句话。"
```

#### 保留分隔符
```python
splitter = CharacterTextSplitter(
    chunk_size=20,
    chunk_overlap=5,
    separator="。",
    keep_separator=True
)
chunks = splitter.split_text(text)
```

**输出：**
```python
[
    '第一句话。第二句话。',
    '句话。第三句话。',  # 重叠部分：句话。
    '句话。第四句话。'
]
```

#### 删除分隔符
```python
splitter = CharacterTextSplitter(
    chunk_size=20,
    chunk_overlap=5,
    separator="。",
    keep_separator=False
)
chunks = splitter.split_text(text)
```

**输出：**
```python
[
    '第一句话第二句话',
    '句话第三句话',  # 重叠部分：句话（没有句号）
    '句话第四句话'
]
```

**对比：**
- 保留分隔符时，overlap 区域有明确的句子边界（`句话。`）
- 删除分隔符时，overlap 区域变成碎片（`句话`），可能语义不清

---

## 不同分隔符的最佳实践

| 分隔符类型 | 推荐设置 | 理由 |
|-----------|---------|------|
| **中文句号** `。` | `keep_separator=True` | 保持句子完整性 |
| **英文句号** `.` | `keep_separator=True` | 同上 |
| **问号/感叹号** `？！` | `keep_separator=True` | 保留语气和情感 |
| **换行符** `\n` | `keep_separator=False` | 换行符通常是格式，非语义 |
| **逗号** `,` | `keep_separator=True` | 保持短语结构 |
| **分号** `;` | 根据场景 | 自然文本用 True，结构化数据用 False |
| **自定义标记** `###` | `keep_separator=False` | 标记通常无语义价值 |


## 总结建议

### ✅ 使用 `keep_separator=True` 的场景：
1. **自然语言文本**（对话、文章、FAQ）
2. **需要保持语义完整性**的场景
3. **面向人类阅读**的输出
4. **训练对话模型**的数据预处理

### ✅ 使用 `keep_separator=False` 的场景：
1. **结构化数据**（CSV、分号分隔的列表）
2. **分隔符无语义价值**（如特殊标记 `###`）
3. **需要干净数据**便于进一步解析

## 3、split by tokens



cl100k_base 是 OpenAI 开发的一种 Token 编码器，用于将文本转换为 Token（模型能理解的最小单位）。核心特点

词汇表大小：约 100,000 个 Token
支持模型：GPT-4、GPT-3.5-turbo、text-embedding-ada-002
多语言支持：优化了中文、日文等非英语语言的编码效率
常用程度：⭐⭐⭐⭐⭐ 最常用

````python
# 1. 导入相关依赖
from langchain_text_splitters import CharacterTextSplitter
import tiktoken  # 用于计算Token数量


# 2. 定义通过Token切割器
text_splitter = CharacterTextSplitter.from_tiktoken_encoder(
    encoding_name="cl100k_base", 
    chunk_size=31,
    chunk_overlap=0,
    separator="。",  # 指定中文句号为分隔符
    keep_separator=False,  # chunk中是否保留分隔符
)
# 3.定义文本
text = "人工智能是一个强大的开发框架。它支持多种语言模型和工具链。今天你吃了吗？我好想吃饭，但是现在不饿 咋办 12345不说话12345不说话12345不说话"

# 4. 开始切割
texts = text_splitter.split_text(text)
print(f"分割后的块数: {len(texts)}")

# 5. 初始化tiktoken编码器（用于Token计数）
encoder = tiktoken.get_encoding("cl100k_base")  # 确保与CharacterTextSplitter的encoding_name一致

# 6. 打印每个块的Token数和内容
for i, chunk in enumerate(texts):
    tokens = encoder.encode(chunk)  # 现在encoder已定义
    print(f"块 {i + 1}: {len(tokens)} Token\n内容: {chunk}\n")
    
# 您设置了 chunk_size=31（每块最多31个token），但第3块有47个token，超出了限制！这是为什么？
# 核心原因：找不到合适的分隔符
# CharacterTextSplitter 的工作流程：
# 
# 尝试按 separator="。" 分割
# 检查每个分割后的片段是否 ≤ chunk_size（31 tokens）
# 如果某个片段 > chunk_size，且没有更小的分隔符可用，就会：
# 
# ⚠️ 保留整个片段（即使超出限制）
# ⚠️ 不做进一步切分
# 
# 关键矛盾：
# 
# 你指定的分隔符：只有 "。"
# 句子3内部没有 "。"，只有 "？"、"，" 和空格
# 因此，无法按 "。" 继续分割
# LangChain 的默认行为：保留整个句子（即使超出 chunk_size）
````

### 增加分隔符来解决

````python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# 使用递归分割器，支持多级分隔符
text_splitter = RecursiveCharacterTextSplitter.from_tiktoken_encoder(
    encoding_name="cl100k_base",
    chunk_size=31,
    chunk_overlap=0,
    separators=["。", "？", "，", " "],  # 多级分隔符
    keep_separator=False
)

text = "人工智能是一个强大的开发框架。它支持多种语言模型和工具链。今天你吃了吗？我好想吃饭，但是现在不饿 咋办 12345不说话12345不说话12345不说话"

chunks = text_splitter.split_text(text)

encoder = tiktoken.get_encoding("cl100k_base")
for i, chunk in enumerate(chunks):
    tokens = len(encoder.encode(chunk))
    print(f"块 {i + 1}: {tokens} tokens - {chunk}\n")
````

## 4、SemanticChunker 语义分块



````python
from langchain_experimental.text_splitter import SemanticChunker
from langchain_openai.embeddings import OpenAIEmbeddings
import os
import dotenv

dotenv.load_dotenv()

# 加载文本
with open("asset/load/09-ai1.txt", encoding="utf-8") as f:
    state_of_the_union = f.read()  #返回字符串

# 获取嵌入模型
os.environ['OPENAI_API_KEY'] = os.getenv("OPENAI_API_KEY1")
os.environ['OPENAI_BASE_URL'] = os.getenv("OPENAI_BASE_URL")
embed_model = OpenAIEmbeddings(
    model="text-embedding-3-large"
)

# 获取切割器
text_splitter = SemanticChunker(
    embeddings=embed_model,
    breakpoint_threshold_type="percentile",#断点阈值类型：字面值["百分位数", "标准差", "四分位距", "梯度"] 选其一
    breakpoint_threshold_amount=65.0 #断点阈值数量 (极低阈值 → 高分割敏感度)
)

# 切分文档
docs = text_splitter.create_documents(texts = [state_of_the_union])
print(len(docs)) # 返回文档数
for doc in docs:
    print('-'*50)
    print(f"🔍 文档 {doc}:")
````
