```python
from langchain_openai import ChatOpenAI
from langchain_core.messages import SystemMessage, HumanMessage

# 创建了一个对话模型
chat_model = ChatOpenAI()

# 在一个消息列表中包含多个问题
messages = [
    SystemMessage(content="你是一个诗人,名字叫李小白"),
    HumanMessage(content="1. 写一首关于春天的诗 2. 你叫什么名字？请分别回答这两个问题。"),
]

# 传入的是消息列表
response = chat_model.invoke(messages)  # 输入消息列表
print(response)
# AIMessage:表示交互完以后，大模型返回的信息
print(type(response))  # <class 'langchain_core.messages.ai.AIMessage'>

```

```python
# 另一种方式：使用批处理方式处理多个请求
batch_messages = [
    [
        SystemMessage(content="你是一个诗人,名字叫李小白"),
        HumanMessage(content="写一首关于春天的诗"),
    ],
    [
        SystemMessage(content="你可以针对用户的提问，进行回复"),
        HumanMessage(content="你叫什么名字？"),
    ]
]

# 使用batch方法一次性获取多个回答
batch_responses = chat_model.batch(batch_messages)
print("\n批量处理多个问题的结果:")
for i, resp in enumerate(batch_responses):
    print(f"问题 {i+1} 的回答:")
    print(resp)
```

# 3、 OpenAI的api的调用演示

非对话模型

```python
from openai import OpenAI
import os

# 从环境变量读取API密钥（推荐安全存储）
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY1"),  #填写自己的api-key
                base_url="https://api.openai-proxy.org/v1") #通过代码示例获取

# 调用Completion接口
response = client.completions.create(
    model="gpt-3.5-turbo-instruct",  # 非对话模型
    prompt="请将以下英文翻译成中文：\n'Artificial intelligence will reshape the future.'",
    max_tokens=100,  # 生成文本最大长度
    temperature=0.7,  # 控制随机性（0-2）
)
# 提取结果
print(response.choices[0].text.strip())
```

对话模型的调用：



```python
from openai import OpenAI

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY1"), #填写自己的api-key
                base_url="https://api.openai-proxy.org/v1")

completion = client.chat.completions.create(
    model="gpt-3.5-turbo", # 对话模型
    messages=[
        {"role": "system", "content": "你是一个乐于助人的智能AI小助手"},
        {"role": "user", "content": "你好，请你介绍一下你自己"}
    ],
    max_tokens=150,
    temperature=0.5
)

print(completion.choices[0].message)
```

使用LangChain，调用对话模型



```python
#0、导包
import os
import dotenv
dotenv.load_dotenv()
from langchain_openai import ChatOpenAI

#1、给两个环境变量赋值
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY1")
os.environ["OPENAI_BASE_URL"] = os.getenv("OPENAI_BASE_URL")

#2、调用ChatOpenAI()返回对话模型
chat_model = ChatOpenAI(
    model="gpt-4o-mini",
    max_tokens=20,#设置返回的文本的最大token值
    temperature=0
)

#3、通过对话模型，调用invoke()
response = chat_model.invoke("什么是LangChain?")
print(response.content)
print(type(response))
```
