# GPT 系列模型结构及创新点总结

## GPT-1

- **模型结构：**
  - 12 层的 Decoder-only Transformer（仅用 Transformer 的解码器部分）
  - 隐层维度 768
  - 注意力头数 12
  - 参数量约 1.17 亿

- **创新点：**
  - 首次将大规模的 Transformer 用于文本的预训练（无监督的语言建模任务）
  - 只用简单的单向语言模型进行预训练（左到右）
  - 预训练后，通过下游任务的微调来适配具体任务

---

## GPT-2

- **模型结构：**
  - 最高版本为 48 层 Decoder-only Transformer
  - 隐层维度 1600
  - 注意力头数 25
  - 最大参数量 15 亿（small/medium/large/XL 四种体量）
  - 仅以自回归（Autoregressive）方法预测下一个 token

- **创新点：**
  - 极大地扩展了模型与数据集的规模
  - 取消下游任务的微调，仅用“零样本/少样本/多样本学习”，只需给模型文字提示即可
  - 展现出“预训练 + 提示（prompt）”的范式强大潜力

---

## GPT-3

- **模型结构：**
  - 96 层 Decoder-only Transformer
  - 隐层维度 12288
  - 注意力头数 96
  - 参数量 1750 亿

- **创新点：**
  - 参数量大幅提升至 1750 亿，数据集和语料规模空前庞大
  - “少样本学习（Few-shot）”表现显著提升
  - 强化“预训练 + prompt 工程”而无需传统的微调
  - 展现出极强的泛化能力和多任务能力

---

## ChatGPT

- **模型结构：**
  - 基于 GPT-3 或 GPT-3.5/4 架构（Decoder-only Transformer）
  - 集成 RLHF（人类反馈强化学习），结构上类似 GPT-3，但训练流程有变

- **创新点：**
  - 首次引入“基于人类反馈的强化学习”（RLHF）进行对话能力强化
  - 增强模型对指令和人类偏好的理解与遵循
  - 对话连贯性和安全性显著提升，支持多轮复杂对话
  - 极大改善在人类对话、问答、辅助等场景下的实用表现



## BERT Encoder-only 与 GPT Decoder-only 的差异

| 特点                   | BERT（Encoder-only）                      | GPT（Decoder-only）                       |
|------------------------|-------------------------------------------|-------------------------------------------|
| 结构                   | 仅包含 Transformer Encoder 堆叠            | 仅包含 Transformer Decoder 堆叠            |
| 输入信息类型           | 双向编码（利用上下文的双向信息）           | 单向编码（仅利用左侧已生成的上下文信息）    |
| 预训练目标             | 掩码语言模型（Masked Language Model, MLM）| 自回归语言模型（Auto-regressive LM）      |
| 下游任务适用           | 适合理解类任务（分类、问答、句子匹配等）   | 适合生成类任务（文本生成、续写等）         |
| 训练时输入遮挡         | 随机掩码部分输入，预测被mask的词           | 仅依赖已生成的token，不能看到未来词         |
| 代表模型               | BERT、RoBERTa、ALBERT等                    | GPT-1、GPT-2、GPT-3、ChatGPT等            |

**简要总结：**
- BERT 的 Encoder-only 结构擅长文本的理解类任务，能够同时关注上文与下文；
- GPT 的 Decoder-only 结构擅长文本生成，仅能看到当前位置之前的内容，流式生成下一个 token。
