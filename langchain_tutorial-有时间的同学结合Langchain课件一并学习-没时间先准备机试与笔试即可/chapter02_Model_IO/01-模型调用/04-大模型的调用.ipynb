```python
import os
import dotenv
from langchain_openai import ChatOpenAI

dotenv.load_dotenv()

os.environ['OPENAI_API_KEY'] = os.getenv("OPENAI_API_KEY1")
os.environ['OPENAI_BASE_URL'] = os.getenv("OPENAI_BASE_URL")

chat_model = ChatOpenAI(
	model="gpt-5-mini",
	temperature=1,
)
```

```python
from langchain_core.messages import SystemMessage, HumanMessage

sys_message = SystemMessage(
    content="我是一个人工智能的助手，我的名字叫小智",
)
human_message = HumanMessage(content="猫王是一只猫吗？")

messages = [sys_message, human_message]

#调用大模型，传入messages
response = chat_model.invoke(messages)
print(response.content)


response1 = chat_model.invoke("你叫什么名字？") #并不会记住名字，因为这是两次对话
print(response1.content)
# print(response.content)
```

```python
from langchain_core.messages import SystemMessage, HumanMessage  # 导入必要的消息类型

sys_message = SystemMessage(  # 创建系统消息
    content="我是一个人工智能的助手，我的名字叫小智",  # 设置系统消息内容
)
human_message = HumanMessage(content="猫王是一只猫吗？")  # 创建第一个人类消息
human_message1 = HumanMessage(content="你叫什么名字？")  # 创建第二个人类消息

messages = [sys_message, human_message, human_message1]  # 将所有消息组合成一个列表

#调用大模型，传入messages
response = chat_model.invoke(messages)  # 调用大模型API，传入消息列表
print(response.content)  # 打印模型的回复内容
```

```python
from langchain_core.messages import SystemMessage, HumanMessage

sys_message = SystemMessage(
    content="我是一个人工智能的助手，我的名字叫小智",
)
human_message = HumanMessage(content="猫王是一只猫吗？")

sys_message1 = SystemMessage(
    content="我可以做很多事情，有需要就找我吧",
)

human_message1 = HumanMessage(content="你叫什么名字？")
#多个 SystemMessage 会被合并理解,会优先回答最新的问题，也就是human_message1
messages = [sys_message, human_message,sys_message1,human_message1]

#调用大模型，传入messages
response = chat_model.invoke(messages)
print(response.content)
```

```python
from langchain_core.messages import SystemMessage, HumanMessage

sys_message = SystemMessage(
    content="我是一个人工智能的助手，我的名字叫小智",
)
human_message = HumanMessage(content="猫王是一只猫吗？")

messages = [sys_message, human_message]

# 另一组
sys_message1 = SystemMessage(
    content="我可以做很多事情，有需要就找我吧",
)

human_message1 = HumanMessage(content="你叫什么名字？")

messages1 = [sys_message1,human_message1]

#调用大模型，传入messages
response = chat_model.invoke(messages)
print(response.content)

response = chat_model.invoke(messages1)
print(response.content)
```

# 多轮对话的体现：

```python
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage

messages = [
    SystemMessage(content="我是一个人工智能助手，我的名字叫小智"),
    HumanMessage(content="人工智能英文怎么说？"),
    AIMessage(content="AI"),
    HumanMessage(content="你叫什么名字"),
]

messages1 = [
    SystemMessage(content="我是一个人工智能助手，我的名字叫小智"),
    HumanMessage(content="很高兴认识你"),
    AIMessage(content="我也很高兴认识你"),
    HumanMessage(content="你叫什么名字"),
]

messages2 = [
    SystemMessage(content="我是一个人工智能助手，我的名字叫小智"),
    HumanMessage(content="很高兴认识你"),
    SystemMessage(content="我也很高兴认识你"),
    HumanMessage(content="你叫什么名字"),
]

messages3 = [
    SystemMessage(content="我是一个人工智能助手，我的名字叫小智"),
    HumanMessage(content="人工智能英文怎么说？"),
    AIMessage(content="AI"),
    SystemMessage(content="我很高兴认识你"),
    HumanMessage(content="你叫什么名字"),
]

chat_model.invoke(messages1)
```
