# message的使用和大模型的调用

## 1、 Message的使用

讲对话模型时提到过： 输入：List[BaseMessage] 或 PromptValue;输出：BaseMessage（其子类AIMessage)


如下提供了各种不同的消息的创建，重点需要掌握和关注的是：SystemMessage、HumanMessage;AIMessage

```python
from langchain_core.messages import HumanMessage, SystemMessage, ChatMessage

sys_message = SystemMessage(content="你是一个人工智能的助手，名字叫小智")
human_message = HumanMessage(content="猫王是一只猫吗？")

messages = [sys_message, human_message]

print(messages)
```

```python
from langchain_core.messages import HumanMessage, SystemMessage

sys_message = SystemMessage(
    content="你是一个人工智能的助手",
    additional_kwargs={"name":"小智"}
)
human_message = HumanMessage(content="猫王是一只猫吗？")

messages = [sys_message, human_message]

print(messages)
```

```python
from langchain_core.messages import HumanMessage, SystemMessage,AIMessage

sys_message = SystemMessage(
    content="你是一个人工智能的助手",
    additional_kwargs={"name":"小智"}
)
human_message = HumanMessage(content="猫王是一只猫吗？")

ai_message = AIMessage(content="猫王不是猫")

messages = [sys_message, human_message,ai_message]

print(messages)
```

```python
from langchain_core.messages import HumanMessage, SystemMessage,AIMessage,ChatMessage

sys_message = SystemMessage(
    content="你是一个人工智能的助手",
    additional_kwargs={"name":"小智"}
)
human_message = HumanMessage(content="猫王是一只猫吗？")

ai_message = AIMessage(content="猫王不是猫")

chat_message = ChatMessage(
    role="waiter",
    content="你是一个服务员，可以提供客户很多的需求"
)

messages = [sys_message, human_message,ai_message,chat_message]

print(messages)
```

# 结合大模型的调用



```python
#0、导包
import os
import dotenv
dotenv.load_dotenv()
from langchain_openai import ChatOpenAI

#1、给两个环境变量赋值
os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY1")
os.environ["OPENAI_BASE_URL"] = os.getenv("OPENAI_BASE_URL")

#2、调用ChatOpenAI()返回对话模型
chat_model = ChatOpenAI(
    model="gpt-5-mini",
    # max_tokens=20,#设置返回的文本的最大token值
    temperature=1  #这里必须是1，mini不支持其他值
)
```

```python
sys_message = SystemMessage(
    content="你是一个人工智能的助手，你的名字叫小智",
)
human_message = HumanMessage(content="猫王是一只猫吗？")

messages = [sys_message, human_message]

#调用大模型，传入messages

response = chat_model.invoke(messages)
print(type(response))  #<class 'langchain_core.messages.ai.AIMessage'>
# print(response)
print(response.content)


response1 = chat_model.invoke("你叫什么名字？") #因为这里和上面是两次交互，所以第二次我们提问，它并不知道自己的名字叫小智
print(response1.content)
# print(response.content)
```

## 2、 大模型的调用

除了invoke()之外，还有哪些方法可以调用大模型呢？


第1波：invoke() / stream()

举例：调用invoke()

```python
import os
import dotenv
from langchain_core.messages import HumanMessage
from langchain_openai import ChatOpenAI

dotenv.load_dotenv()

os.environ['OPENAI_API_KEY'] = os.getenv("OPENAI_API_KEY1")
os.environ['OPENAI_BASE_URL'] = os.getenv("OPENAI_BASE_URL")

#初始化大模型
chat_model = ChatOpenAI(model="gpt-5-mini",temperature=1)

# 创建消息
messages = [HumanMessage(content="你好，请介绍一下自己")]

# 非流式调用LLM获取响应
response = chat_model.invoke(messages)

# 打印响应内容
print(response)
```

举例：调用stream()


```python
import os
import dotenv
from langchain_core.messages import HumanMessage
from langchain_openai import ChatOpenAI

dotenv.load_dotenv()

os.environ['OPENAI_API_KEY'] = os.getenv("OPENAI_API_KEY1")
os.environ['OPENAI_BASE_URL'] = os.getenv("OPENAI_BASE_URL")

#初始化大模型
chat_model = ChatOpenAI(
    model="gpt-5-mini",
    streaming=True,
    temperature=1
)

# 创建消息
messages = [HumanMessage(content="langchain是什么？内部包含哪些知识？")]

# 流式调用LLM获取响应
chunks = chat_model.stream(messages)

for chunk in chunks:
    print(chunk.content,end="",flush=True)
```

第2波：batch()：批量调用的使用

```python
import os
import dotenv
from langchain_core.messages import HumanMessage, SystemMessage
from langchain_openai import ChatOpenAI

dotenv.load_dotenv()

os.environ['OPENAI_API_KEY'] = os.getenv("OPENAI_API_KEY1")
os.environ['OPENAI_BASE_URL'] = os.getenv("OPENAI_BASE_URL")

# 初始化大模型
chat_model = ChatOpenAI(model="gpt-4o-mini")

messages1 = [SystemMessage(content="你是一位乐于助人的智能小助手"),
             HumanMessage(content="请帮我介绍一下什么是机器学习"), ]

messages2 = [SystemMessage(content="你是一位乐于助人的智能小助手"),
             HumanMessage(content="请帮我介绍一下什么是AIGC"), ]

messages3 = [SystemMessage(content="你是一位乐于助人的智能小助手"),
             HumanMessage(content="请帮我介绍一下什么是大模型技术"), ]

messages = [messages1, messages2, messages3]

# 调用batch
response = chat_model.batch(messages)
# response = chat_model.invoke(messages)  #调用失败

print(response)
```

第3波：同步方法与异步方法



```python
import time

def call_model():
    # 模拟同步API调用
    print("开始调用模型...")
    time.sleep(5)  # 模拟调用等待,单位：秒
    print("模型调用完成。")

def perform_other_tasks():
    # 模拟执行其他任务
    for i in range(5):
        print(f"执行其他任务 {i + 1}")
        time.sleep(1)  # 单位：秒

def main():
    start_time = time.time()

    call_model()
    perform_other_tasks()

    end_time = time.time()

    total_time = end_time - start_time
    return f"总共耗时：{total_time}秒"

# 运行同步任务并打印完成时间
main_time = main()
print(main_time)
```

```python
import asyncio  # 导入asyncio模块，用于支持异步编程
import time  # 导入time模块，用于计时功能

#同步： synchronized  # 注释说明同步的英文
#异步： asynchronized  # 注释说明异步的英文

async def async_call(llm):  # 定义一个异步函数，模拟调用语言模型
    await asyncio.sleep(5)  # 模拟异步操作，等待5秒钟
    print("异步调用完成")  # 打印调用完成的消息

async def perform_other_tasks():  # 定义一个异步函数，模拟执行其他任务
    await asyncio.sleep(5)  # 模拟异步操作，等待5秒钟
    print("其他任务完成")  # 打印任务完成的消息

async def run_async_tasks():  # 定义一个异步函数，用于运行所有异步任务
    start_time = time.time()  # 记录开始时间

    await asyncio.gather(  # 使用gather同时运行多个协程
        async_call(None),  # 示例调用，替换None为模拟的LLM对象
        perform_other_tasks()  # 同时执行其他任务
    )

    end_time = time.time()  # 记录结束时间
    return f"总共耗时：{end_time - start_time}秒"  # 返回总耗时

# # 正确运行异步任务的方式  # 注释说明这是在普通Python脚本中运行的方式，就是后缀是py文件中
# if __name__ == "__main__":  # 注释说明这是主程序入口点
# #     # 使用 asyncio.run() 来启动异步程序  # 注释说明使用asyncio.run()启动异步程序
#     result = asyncio.run(run_async_tasks())  # 注释说明运行异步任务并获取结果
#     print(result)  # 注释说明打印结果


# 在 Jupyter 单元格中直接调用  # 注释说明这是在Jupyter中运行的方式
result = await run_async_tasks()  # 在Jupyter中直接使用await关键字运行异步函数,红色波浪线不用管
print(result)  # 打印运行结果
```
