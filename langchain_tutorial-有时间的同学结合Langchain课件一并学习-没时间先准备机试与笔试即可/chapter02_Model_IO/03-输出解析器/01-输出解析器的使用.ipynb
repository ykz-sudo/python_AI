# Model IO之输出解析器的使用

## 1、StrOutputParser

```python
from langchain_core.output_parsers import StrOutputParser

#1、创建大模型
from langchain_core.messages import HumanMessage, SystemMessage
from langchain_core.output_parsers import StrOutputParser

import os
import dotenv
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI

dotenv.load_dotenv()

os.environ['OPENAI_API_KEY'] = os.getenv("OPENAI_API_KEY1")
os.environ['OPENAI_BASE_URL'] = os.getenv("OPENAI_BASE_URL")

chat_model = ChatOpenAI(model="gpt-4o-mini")
#2、提供提示词模板，并填充变量
prompt_template = ChatPromptTemplate.from_messages([
    ("system","你是一个人工智能的助手"),
    ("human","{input}")
    # SystemMessage(content="你是一个人工智能的助手"),
    # HumanMessage(content="{input}")
])

# prompt_value = prompt_template.invoke({"input":"解释一下什么是大模型？"})
prompt_value = prompt_template.format(input="解释一下什么是大模型？")
# print(prompt_value)

#3、大模型调用提示词模板，返回一个结果
message = chat_model.invoke(prompt_value)

print(message)
# print(type(message))
print('-'*50)
print(message.content)
print('-'*50)
# #4、创建字符串解析器
parser = StrOutputParser()
#
# #5、将大模型返回的结果传入解析器，得到字符串类型的结果
str = parser.invoke(message)
print(str)
print(type(str))
```

```python
from langchain_core.output_parsers import JsonOutputParser
from langchain_core.prompts import ChatPromptTemplate

chat_model = ChatOpenAI(model="gpt-4o-mini")

chat_prompt_template = ChatPromptTemplate.from_messages([
    ("system","你是一个靠谱的{role},请使用{format_instruction}格式回复"),
    ("human","{question}")
])


parser = JsonOutputParser()

prompt = chat_prompt_template.format_messages(
    role="人工智能专家",
    question="人工智能用英文怎么说？",
    format_instruction=parser.get_format_instructions()
)
result = chat_model.invoke(prompt)
# print(result)
# print(type(result))

parser.invoke(result)
```

其中，parser.get_format_instructions() 返回一个字符串，内容就是要求用户返回一个JSON对象，所以，我们在提示词中，就直接调用此变量即可，不用自己用语言说明了。

```python
# 引入依赖包
from langchain_core.output_parsers import JsonOutputParser
from langchain_core.prompts import PromptTemplate

# 初始化语言模型
chat_model = ChatOpenAI(model="gpt-4o-mini")



# 定义Json解析器
parser = JsonOutputParser()

# 定义提示词模版
# 注意，提示词模板中需要部分格式化解析器的格式要求format_instructions
prompt = PromptTemplate(
    template="回答用户的查询.\n{format_instructions}\n{query}\n",
    input_variables=["query"],
    partial_variables={"format_instructions": parser.get_format_instructions()},
)

#使用以前的方式：
joke_query = "告诉我一个笑话。"
prompt_value = prompt.invoke({"query":joke_query})
message = chat_model.invoke(prompt_value)
json_output = parser.invoke(message)
# print(json_output)
# print(type(json_output))

# LCEL,使用|
chain = prompt | chat_model | parser
json_output1 = chain.invoke({"query":joke_query})
print(json_output1)
print(type(json_output1))
```

## 4、 CommaSeparatedListOutputParser (了解)

```python
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI
from langchain.output_parsers import CommaSeparatedListOutputParser

# 初始化语言模型
chat_model = ChatOpenAI(model="gpt-4o-mini")

# 创建解析器
output_parser = CommaSeparatedListOutputParser()

# 创建LangChain提示模板
chat_prompt = PromptTemplate.from_template(
    "生成5个关于{text}的列表.\n\n{format_instructions}",
    partial_variables={
    "format_instructions": output_parser.get_format_instructions()
    })

# 方式1：
# prompt = chat_prompt.invoke({"text":"心情"})
# message = chat_model.invoke(prompt)
# output_parser.invoke(message)

# 方式2：
chain = chat_prompt | chat_model | output_parser
res = chain.invoke({"text": "电影"})
print(res)
# print(type(res))
```
