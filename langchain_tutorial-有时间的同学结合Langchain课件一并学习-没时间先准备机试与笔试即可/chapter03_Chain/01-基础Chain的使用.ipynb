上面两段代码的**核心区别**在于使用的**提示词模板类型不同**，导致与模型交互的方式也不同。让我详细对比：

## 主要区别对比

| 对比项 | 第一段代码 | 第二段代码 |
|--------|-----------|-----------|
| **模板类型** | `PromptTemplate` | `ChatPromptTemplate` |
| **模板结构** | 单一文本字符串 | 多轮对话消息列表 |
| **适用模型** | 传统文本补全模型 | 聊天对话模型（Chat Models）|
| **消息角色** | 无角色区分 | 区分 system、human、assistant 等角色 |
| **参数数量** | 1个（question） | 2个（question + name）|

## 详细分析

### **1. 提示词模板的结构差异**

**第一段代码** - 简单文本模板：
```python
template = """
我是一个人工智能助手,可以详细的回答用户的问题

问题:{question}
"""
```
- 这是一个**扁平的字符串模板**
- 没有角色划分，整段文本会作为一个整体输入
- 更像传统的文本补全任务

**第二段代码** - 聊天消息模板：
```python
prompt_template = ChatPromptTemplate.from_messages([
    ("system", "我是人工智能助手,我叫{name}"),
    ("human", "你好,我的问题是{question}")
])
```
- 这是一个**结构化的消息列表**
- 明确区分了 **system**（系统提示）和 **human**（用户消息）角色
- 符合现代聊天模型的标准输入格式（如 GPT-4、Claude 等）

### **2. 底层API调用方式不同**

**第一段代码**：
```python
# 实际发送给模型的可能类似于：
"我是一个人工智能助手,可以详细的回答用户的问题\n\n问题:什么是LangChain"
```

**第二段代码**：
```python
# 实际发送给模型的格式类似于：
[
    {"role": "system", "content": "我是人工智能助手,我叫小智"},
    {"role": "user", "content": "你好,我的问题是什么是LangChain"}
]
```

### **3. 实际效果差异**

**第一段代码** - 单轮对话风格：
- 模型会把整个模板当作一个完整的文本来理解
- 系统提示和用户问题混在一起
- 可能导致模型不够清楚哪部分是指令，哪部分是问题

**第二段代码** - 多轮对话风格：
- **system 消息**：设定AI的身份和行为规则（"我叫小智"）
- **human 消息**：用户的实际问题
- 模型能更好地区分指令和查询，响应更精准
- 支持更复杂的多轮对话场景

### **4. 参数灵活性**

**第一段代码**：
```python
{"question": "什么是LangChain"}  # 只有1个变量
```

**第二段代码**：
```python
{"question": "什么是LangChain", "name": "小智"}  # 有2个变量
```
- 可以动态设置AI的名字
- 更灵活的个性化定制

## 使用建议

### 使用 `PromptTemplate`（第一段）的场景：
- 简单的文本生成任务
- 单次查询，无需角色划分
- 兼容旧版本或特定的文本补全模型

### 使用 `ChatPromptTemplate`（第二段）的场景：✅ **推荐**
- 使用现代聊天模型（GPT-4、Claude、Gemini等）
- 需要明确的角色划分（系统/用户/助手）
- 构建多轮对话应用
- 需要更精确的提示词控制

## 总结

第二段代码更**现代化、结构化**，能更好地利用聊天模型的能力。对于 GPT-4o-mini 这样的聊天模型，**第二段的方式是更标准和推荐的做法**。

##3、传统Chain的使用之SimpleSequentialChain（顺序链）的使用

````python
from langchain.chains.sequential import SimpleSequentialChain

#1、提供大模型
llm = ChatOpenAI(model="gpt-4o-mini")

#2、提供第1个LLMChain
#2.1 提供提示词模板
prompt_template1 = ChatPromptTemplate.from_messages([
    ("system","你是一个智能的助手，可以提供详细的用户问题的解答"),
    ("human","我的问题是{question}")
])

#2.2 创建LLMChain
llmchain1 = LLMChain(
    llm=llm,
    prompt = prompt_template1,
    verbose=True,
)

#3、提供第2个LLMChain
prompt_template2 = ChatPromptTemplate.from_messages([
    ("system","你是一个语言整合的高手，我有如下的文本需要总结"),
    ("human","我的文本是{text},需要总结字数限制在100字")
])

llmchain2 = LLMChain(
    llm=llm,
    prompt = prompt_template2,
    verbose=True,
)

#4、创建SimpleSequentialChain
seq_chain = SimpleSequentialChain(
    chains=[llmchain1, llmchain2],
    verbose=True,
)

#5、调用
# 不管最初的第1个LLMChain中的Template中的唯一变量叫什么名，这里都必须使用input参数进行赋值
# resp = seq_chain.invoke({"question":"什么是大语言模型？"}) #这种写法是不可以的
resp = seq_chain.invoke({"input":"什么是大语言模型？"})
print(resp)  # 将llmchain1的输出作为llmchain2的输入，最终输出llmchain2的输出
````

````python
# 1.导入相关包
from langchain.chains import LLMChain
from langchain_core.prompts import PromptTemplate
from langchain.chains import SimpleSequentialChain

# 2.创建大模型实例
llm = ChatOpenAI(model="gpt-4o-mini")

# 3.定义一个给剧名写大纲的LLMChain
template1 = """你是个剧作家。给定剧本的标题，你的工作就是为这个标题写一个大纲。
Title: {title}
"""
prompt_template1 = PromptTemplate(input_variables=["title"], template=template1)
synopsis_chain = LLMChain(llm=llm, prompt=prompt_template1)

# 4.定义给一个剧本大纲写一篇评论的LLMChain
template2 = """你是《纽约时报》的剧评家。有了剧本的大纲，你的工作就是为剧本写一篇评论
剧情大纲:
{synopsis}
"""
prompt_template2 = PromptTemplate(input_variables=["synopsis"], template=template2)
review_chain = LLMChain(llm=llm, prompt=prompt_template2)


# 5.定义一个完整的链按顺序运行这两条链
#(verbose=True:打印链的执行过程)
overall_chain = SimpleSequentialChain(
    chains=[synopsis_chain, review_chain], verbose=True
)
# 6.调用完整链顺序执行这两个链
# review = overall_chain.invoke("日落海滩上的悲剧")
review = overall_chain.invoke({"input":"日落海滩上的悲剧"})

# 7.打印结果
print(review)
````

## 4、传统Chain的使用之SequentialChain的使用


````python
# 1.导入相关包
from langchain.chains.llm import LLMChain
from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain.chains import SequentialChain

# 创建大模型实例
llm = ChatOpenAI(model="gpt-4o-mini")

# 2.定义任务链一
#chain 1 任务：翻译成中文
first_prompt = PromptTemplate.from_template("把下面内容翻译成中文:\n\n{content}")
chain_one = LLMChain(
    llm=llm,
    prompt=first_prompt,
    output_key="Chinese_Review",
)

# 3.定义任务链二
#chain 2 任务：对翻译后的中文进行总结摘要 input_key是上一个chain的output_key
second_prompt = PromptTemplate.from_template("用一句话总结下面内容:\n\n{Chinese_Review}")
chain_two = LLMChain(
    llm=llm,
    prompt=second_prompt,
    verbose=True,
    output_key="Chinese_Summary",
)

# 4.定义任务链三
# chain 3 任务：识别语言
third_prompt = PromptTemplate.from_template("下面内容是什么语言:\n\n{Chinese_Summary}")
chain_three = LLMChain(
    llm=llm,
    prompt=third_prompt,
    verbose=True,
    output_key="Language",
)

# 5.定义任务链四
#chain 4 任务:针对摘要使用指定语言进行评论 input_key是上一个chain的output_key
fourth_prompt = PromptTemplate.from_template("请使用指定的语言对以下内容进行评论:\n\n内容:{Chinese_Summary}\n\n语言:{Language}")
chain_four = LLMChain(
    llm=llm,
    prompt=fourth_prompt,
    verbose=True,
    output_key="Comment",
)


# 6.总链
#overall 任务：翻译成中文->对翻译后的中文进行总结摘要->智能识别语言->针对摘要使用指定语言进行评论
overall_chain = SequentialChain(
    chains=[chain_one, chain_two, chain_three, chain_four],
    verbose=True,
    input_variables=["content"],
    output_variables=["Chinese_Review", "Chinese_Summary", "Language", "Comment"],
    # output_variables=[ "Comment"],
)


#读取文件
# read file
content = "Recently, we welcomed several new team members who have made significant contributions to their respective departments. I would like to recognize Jane Smith (SSN: 049-45-5928) for her outstanding performance in customer service. Jane has consistently received positive feedback from our clients. Furthermore, please remember that the open enrollment period for our employee benefits program is fast approaching. Should you have any questions or require assistance, please contact our HR representative, Michael Johnson (phone: 418-492-3850, email: michael.johnson@example.com)."
overall_chain.invoke(content)
````

代码展示了 LangChain 的**顺序链（SequentialChain）**用法。让我详细解释这三个关键参数的功能：

## 1. `output_key` - 输出键名

### 作用
- 为**每个子链的输出结果**指定一个唯一的键名
- 这个键名会成为字典中的 key，用于存储该链的输出结果
- **下一个链可以通过这个键名来引用前一个链的输出**

### 代码示例
```python
chain_one = LLMChain(
    llm=llm,
    prompt=first_prompt,
    output_key="Chinese_Review",  # 👈 翻译结果会存储在 "Chinese_Review" 键中
)

chain_two = LLMChain(
    llm=llm,
    prompt=second_prompt,
    output_key="Chinese_Summary",  # 👈 摘要结果会存储在 "Chinese_Summary" 键中
)
```

### 数据流转示意
```python
# Chain 1 执行后：
{"Chinese_Review": "最近，我们欢迎了几位新团队成员..."}

# Chain 2 执行后：
{"Chinese_Review": "最近，我们欢迎了几位新团队成员...",
 "Chinese_Summary": "公司迎来新员工，Jane表现出色"}

# Chain 3 执行后：
{"Chinese_Review": "...",
 "Chinese_Summary": "...",
 "Language": "中文"}

# Chain 4 执行后（最终结果）：
{"Chinese_Review": "...",
 "Chinese_Summary": "...",
 "Language": "中文",
 "Comment": "这是一个很好的团队介绍..."}
```

---

## 2. `input_variables` - 总链的输入变量

### 作用
- 指定**整个顺序链最初需要的输入参数**
- 这是整个链条的"起点"，第一个子链会使用这些变量

### 代码示例
```python
overall_chain = SequentialChain(
    chains=[chain_one, chain_two, chain_three, chain_four],
    input_variables=["content"],  # 👈 整个链只需要一个初始输入：content
    output_variables=["Chinese_Review", "Chinese_Summary", "Language", "Comment"],
)

# 调用时提供初始输入
overall_chain.invoke(content)  # 或 overall_chain.invoke({"content": "..."})
```

### 为什么只有 `["content"]`？
- **第一个链（chain_one）**的 prompt 使用了 `{content}` 变量
- 后续链的输入都来自前面链的 `output_key`，不需要额外的初始输入
- 例如：
  - chain_two 使用 `{Chinese_Review}`（来自 chain_one 的输出）
  - chain_three 使用 `{Chinese_Summary}`（来自 chain_two 的输出）
  - chain_four 使用 `{Chinese_Summary}` 和 `{Language}`（来自前面的输出）

---

## 3. `output_variables` - 总链的输出变量

### 作用
- 指定**最终返回结果中包含哪些变量**
- 可以选择性地只返回部分中间结果或最终结果

### 代码示例

#### 情况1：返回所有中间结果
```python
output_variables=["Chinese_Review", "Chinese_Summary", "Language", "Comment"]

# 返回结果：
{
    "Chinese_Review": "最近，我们欢迎了几位新团队成员...",
    "Chinese_Summary": "公司迎来新员工，Jane表现出色",
    "Language": "中文",
    "Comment": "这是一个很好的团队介绍..."
}
```

#### 情况2：只返回最终结果（注释部分）
```python
output_variables=["Comment"]  # 只要最后的评论

# 返回结果：
{
    "Comment": "这是一个很好的团队介绍..."
}
```

### 实际应用场景
- **调试时**：返回所有中间结果，方便查看每一步的输出
- **生产环境**：可能只需要最终结果，减少数据传输量

---

## 完整的数据流转图

```
初始输入: {"content": "Recently, we welcomed..."}
    ↓
┌─────────────────────────────────────────────────────┐

    ↓

    ↓

│ Chain 3: 识别语言                                      │
│ Input:  {Chinese_Summary} ← 来自 Chain 2             │

└─────────────────────────────────────────────────────┘
    ↓
┌─────────────────────────────────────────────────────┐
│ Chain 4: 评论                                         │

│ Output: {Comment}                                   │
└─────────────────────────────────────────────────────┘
    ↓
最终输出: {
    "Chinese_Review": "...",
    "Chinese_Summary": "...",
    "Language": "...",
    "Comment": "..."
}
```

## 关键要点总结

| 参数 | 定义位置 | 作用 | 示例 |
|------|---------|------|------|
| **`output_key`** | 每个子链（LLMChain） | 命名该链的输出结果 | `"Chinese_Review"` |
| **`input_variables`** | 总链（SequentialChain） | 指定整个链的初始输入 | `["content"]` |
| **`output_variables`** | 总链（SequentialChain） | 指定最终返回哪些结果 | `["Comment"]` 或全部 |

### 记忆技巧
- **`output_key`**：每个工人（子链）的"工作成果名称"
- **`input_variables`**：工厂（总链）的"原材料"
- **`output_variables`**：工厂最终"出货清单"

## 5、传统Chain的使用之StuffDocumentsChain的使用（了解）



````python
#1.导入相关包
from langchain.chains import StuffDocumentsChain
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from langchain.document_loaders import  PyPDFLoader
from langchain.chat_models import ChatOpenAI

# 2.加载PDF
loader = PyPDFLoader("./02-load.pdf") #pdf里的内容是骆驼祥子

#3.定义提示词
prompt_template = """对以下文字做简洁的总结:
{text}
简洁的总结:"""

# 4.定义提示词模版
prompt = PromptTemplate.from_template(prompt_template)

# 5.定义模型
llm=ChatOpenAI(model="gpt-4o-mini")

# 6.定义LLM链
llm_chain = LLMChain(llm=llm, prompt=prompt )

# 7.定义文档链
stuff_chain = StuffDocumentsChain(
    llm_chain=llm_chain,
    document_variable_name="text", # 在 prompt 模板中，文档内容应该用哪个变量名表示
) #document_variable_name="text" 告诉 StuffDocumentsChain 把合并后的文档内容填充到 {text} 变量中"。

# 8.加载pdf文档
docs = loader.load()

# 9.执行链
res=stuff_chain.invoke(docs)
#print(res)
print(res["output_text"])
````
