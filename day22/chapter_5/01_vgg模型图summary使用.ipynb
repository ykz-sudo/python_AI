```python
# 解释VGG11结构
print("VGG11结构说明：")
print("""
VGG11是一种经典的卷积神经网络结构，主要包含三个部分：
1. 特征提取部分（features）: 由8层卷积层和ReLU激活函数及5层最大池化（Max Pooling）组成。
   - 前两层为卷积与池化，通道分别从3->64->128。
   - 中间三组为多卷积和池化，通道分别为128->256->512->512，每组均为两层卷积加一次池化。
   - 所有卷积核大小均为3x3，步幅为1，填充为1，池化为2x2。

2. 自适应平均池化层（avgpool）: 将卷积后的特征图尺寸变为7x7，为全连接做准备。

3. 分类器部分（classifier）: 3个全连接层，前两个输出4096，最后输出1000（分类数），中间有ReLU和Dropout防止过拟合。
   - Linear(25088->4096) -> ReLU -> Dropout
   - Linear(4096->4096)  -> ReLU -> Dropout
   - Linear(4096->1000)  作为最终输出（ImageNet有1000类）

VGG11整体特点是结构简单、所有卷积核均为3x3，池化为2x2，便于堆叠。参数量约1.33亿，代表了深层网络设计的经典思想。
""")



# # 检查Graphviz的'dot'可执行文件是否在系统PATH中，给出友好提示
# import shutil
# if shutil.which("dot") is None:
#     print("警告：未检测到Graphviz的'dot'可执行文件，请先安装Graphviz并将其添加到系统PATH环境变量中，否则无法生成模型结构可视化图片。")
#     print("安装方法示例：")
#     print("  - Windows: 请前往 https://graphviz.gitlab.io/_pages/Download/Download_windows.html 下载Graphviz，安装后将其bin目录添加到环境变量PATH。")
#     print("  - Linux/macOS: 可使用 'sudo apt install graphviz' 或 'brew install graphviz' 命令安装。")


# 导入torchvision中的models模块，用于加载各种预训练模型
import torchvision.models as models

# 再次导入models（冗余，可删除，但为说明保留）

# 加载预训练的VGG11模型，参数pretrained=True表示使用在ImageNet上训练好的权重
vgg11 = models.vgg11(pretrained=True)

# 打印VGG11模型的完整结构信息（包含各层配置与参数等）
print(vgg11)

# 计算并打印VGG11模型中参数的总数量
total_params = sum(p.numel() for p in vgg11.parameters())
# 这行代码使用了Python的f-string格式化输出，{total_params:,}中的','会以千位分隔符显示数字
print(f"vgg11模型的总参数量: {total_params:,}")  

# 导入可视化及相关操作所需的库
import torch  # PyTorch框架核心库
from torchviz import make_dot  # 用于可视化计算图
from torch.autograd import Variable  # 用于封装tensor，支持自动求导
import matplotlib.pyplot as plt  # 用于画图
import io  # 用于I/O操作（本代码未用到，仅导入）
from PIL import Image  # 用于图片打开和处理
from torchsummary import summary  # 用于打印模型详细参数信息
import numpy as np  # 数值计算库，用于数据处理

# 创建一个示例输入，形状为(1, 3, 224, 224)，模拟单张3通道224x224图片

# 你说的“这种对象”指的是Variable对象
# 在早期PyTorch（1.x之前）必须将tensor用Variable封装才能反向传播
# 新版PyTorch中，Tensor本身已具备自动求导功能，Variable其实已经等效于Tensor，可以直接用Tensor
# 此处用Variable(torch.randn(1, 3, 224, 224))和直接用torch.randn(1, 3, 224, 224)一样
# 只是习惯写法（或代码兼容老版本），直接用tensor即可

x = Variable(torch.randn(1, 3, 224, 224))

# 将输入数据传入模型，获得前向传播输出（VGG11的预测结果）
y = vgg11(x)

print(f"输入形状: {x.shape}")
print(f"输出形状: {y.shape}")

# # 设置matplotlib支持中文字体，避免中文乱码
# from matplotlib import rcParams
# rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei', 'Arial Unicode MS']  # 添加常用中文字体以支持显示
# rcParams['axes.unicode_minus'] = False  # 保证负号可以正常显示

# # 调用make_dot生成VGG11的计算图（包含各操作节点和权重参数）
# dot = make_dot(y, params=dict(vgg11.named_parameters()))

# # 设置保存生成图片的格式为PNG，并保存到当前路径下，文件名为vgg11_structure.png
# dot.format = 'png'
# dot.render("vgg11_structure", cleanup=True)

# # 尝试读取并显示刚才保存的网络结构图片，如有异常则给出提示
# try:
#     img = Image.open("vgg11_structure.png")  # 打开图片文件
#     plt.figure(figsize=(15, 20),dpi=300)  # 设置图片显示的尺寸和分辨率
#     plt.imshow(np.array(img))  # 显示图片内容
#     plt.axis('off')  # 不显示坐标轴
#     plt.title("VGG11模型结构图")  # 设置标题
#     plt.show()  # 显示图片
# except:
#     # 捕获异常，如图片无法打开或torchviz未安装等
#     print("无法显示图像，请检查torchviz是否正确安装")

# 使用torchsummary.summary函数打印VGG11模型的详细参数信息，包括输入输出尺寸、每层参数数目等
print("VGG11模型的详细参数信息:")
summary(vgg11, (3, 224, 224), device="cpu")
```
