## 缩放点积注意力（Scaled Dot-Product Attention）

缩放点积注意力的计算流程如下：

1. 首先，将输入序列分别通过三个不同的线性变换，得到查询向量 (Q)、键向量 (K)、值向量 (V)。
2. 计算查询与键的点积，得到注意力分数矩阵：$\mathrm{Attention}(Q, K, V) = \mathrm{softmax}\left(\frac{Q K^\top}{\sqrt{d_k}}\right)V$
   - 其中，$d_k$ 是键向量的维度，对点积结果进行缩放（除以 $\sqrt{d_k}$），避免数值过大导致梯度消失或爆炸。
3. 使用 softmax 将注意力分数归一化，得到每个值的权重。
4. 最终输出加权求和值。

## 多头注意力（Multi-Head Attention）

多头注意力原理为：

1. 先将输入分别映射为多个不同的查询、键和值向量，通常有 $h$ 个头。
2. 对每一组（头）独立地进行缩放点积注意力计算，捕捉多种语义特征。
3. 将所有头的输出拼接（concatenate）起来，再通过一个线性变换，得到最终的多头注意力输出。
   - 公式表示：
   $$
   \mathrm{MultiHead}(Q,K,V) = \mathrm{Concat}(\mathrm{head}_1, ..., \mathrm{head}_h)W^O \\
   \text{其中}~ \mathrm{head}_i = \mathrm{Attention}(QW_i^Q, KW_i^K, VW_i^V)
   $$

## 层归一化（Layer Normalization）

层归一化的原理为：

1. 针对每一个样本，对输入的所有特征维度做归一化处理。
2. 计算所有特征的均值 $\mu$ 和方差 $\sigma^2$：
   $$
   \mu = \frac{1}{H} \sum_{i=1}^{H} x_i \\
   \sigma^2 = \frac{1}{H} \sum_{i=1}^{H} (x_i - \mu)^2
   $$
3. 将输入归一化为 $y_i = \frac{x_i - \mu}{\sqrt{\sigma^2 + \epsilon}}$
4. 最后进行仿射变换：$out_i = \gamma y_i + \beta$
   - 其中 $\gamma, \beta$ 是可学习参数。

层归一化的优点是缓解了深层网络中的内部协变量偏移，提高模型的稳定性和收敛速度。
