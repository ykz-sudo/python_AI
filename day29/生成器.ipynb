```python

# 什么是生成器？
# 生成器（Generator）是 Python 中一种特殊的迭代器，用于惰性产生数据。
# 与普通函数不同，生成器用 yield 语句来返回一个值，并且在下次迭代时从上次挂起的地方继续执行。
# 生成器的好处是可以节省内存，适合处理大量数据或无限序列，因为不会一次性把所有数据加载到内存中。


# 迭代器和可迭代对象的关系
# 在Python中，"可迭代对象"（iterable）是指实现了__iter__()方法的对象，可以被for循环遍历。
# 常见的可迭代对象有：list, tuple, str, dict, set等。
# 通过内置的iter()函数，可以从一个可迭代对象获得一个"迭代器"（iterator）。
# "迭代器"（iterator）是实现了__next__()和__iter__()方法的对象。
# 迭代器可以用next()不断取出下一个元素，直到抛出StopIteration异常为止。
# 可迭代对象本身不一定是迭代器，但迭代器一定是可迭代对象。
#
# 举例说明：
my_list = [10, 20, 30]
my_iter = iter(my_list)  # my_iter是一个迭代器
print(next(my_iter))     # 输出10
print(next(my_iter))     # 输出20
print(next(my_iter))     # 输出30
# print(next(my_iter))   # 再调用会抛出StopIteration异常

# 更正式地看：
from collections.abc import Iterable, Iterator

print(isinstance(my_list, Iterable))  # True，可迭代对象
print(isinstance(my_list, Iterator))  # False，不是迭代器

print(isinstance(my_iter, Iterable))  # True，迭代器本身也是可迭代的
print(isinstance(my_iter, Iterator))  # True，是迭代器

# 生成器是迭代器（generator是iterator），同时也是可迭代对象。
# 所以生成器既满足可迭代对象（实现了__iter__）的协议，也满足迭代器（实现了__next__）的协议。
def simple_gen():
    yield 1

g = simple_gen()
print(isinstance(g, Iterator))     # True
print(isinstance(g, Iterable))     # True

# 那么for循环中使用迭代器、可迭代对象和生成器的区别是什么？for循环的底层原理如何？

# 1. for循环支持“可迭代对象”，而不仅仅是迭代器。例如list、str、dict、生成器对象等都可以用于for循环。
# 2. for循环内部会自动调用iter()获取可迭代对象的迭代器，然后不断调用next()取值，直到遇到StopIteration为止。
# 3. 只要对象实现了__iter__（可迭代），都可以参加for循环；对象本身是迭代器也可以直接for循环。
# 4. 生成器也是迭代器，因此可以直接用于for，且每次遇到yield就暂停执行，省内存。

# "for ... in ..."底层原理（伪代码）如下：
#
# it = iter(obj)           # 获得迭代器
# while True:
#     try:
#         item = next(it)  # 取下一个元素
#     except StopIteration:
#         break
#     <循环体代码块>

# 下面用实例演示三种“for”循环对象：

# 1. 可迭代对象 - 列表
lst = [100, 200, 300]
for x in lst:
    print(x)  # 支持for循环，for隐式调用iter(lst)和next()

# 2. 迭代器
it = iter(lst)  # 获得迭代器
for x in it:
    print(x)    # 直接在for里使用迭代器对象

# 3. 生成器
def gen_example():
    yield "A"
    yield "B"
for x in gen_example():
    print(x)    # 生成器对象本质上也是迭代器，支持for

# 总结：
# - for循环的对象可以是任何可迭代对象或者迭代器、生成器。
# - for循环统一转为：获取迭代器，不断next()，直到StopIteration。



def counter(n):
    """
    这是一个生成器函数，它每次产生一个从0到n的整数。
    counter(3)会依次产生0, 1, 2, 3。
    """
    i = 0
    while i <= n:
        yield i  # yield语句会返回i，并在下次迭代时从这里继续
        i += 1

# 使用生成器
my_iter = counter(10)  # 此时my_iter是一个生成器对象
for i in my_iter:
    print(i)

# 也可以用next()函数手动获取下一个值，直到抛出StopIteration异常
# print(next(my_iter))

```
