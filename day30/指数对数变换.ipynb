```python
import torch

import math

 

def positional_encoding_weights(emb_size):

  # 计算底数和指数

  base = math.log(10000)

  exponent = -5 / emb_size #幂次

  

  # 应用指数函数

  weight = torch.exp(torch.tensor(base*exponent))   

  return weight

 

emb_size = 10  # 举例，可以替换成你需要的大小

weight = positional_encoding_weights(emb_size)

print(weight)
```
