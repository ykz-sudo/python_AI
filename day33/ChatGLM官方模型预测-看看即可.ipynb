# ChatGLM-6B 介绍


ChatGLM-6B 是一个开源的、支持中英双语的对话语言模型，基于 [General Language Model (GLM)](https://github.com/THUDM/GLM) 架构，具有 62 亿参数。结合模型量化技术，用户可以在消费级的显卡上进行本地部署（INT4 量化级别下最低只需 6GB 显存）。
ChatGLM-6B 使用了和 ChatGPT 相似的技术，针对中文问答和对话进行了优化。经过约 1T 标识符的中英双语训练，辅以监督微调、反馈自助、人类反馈强化学习等技术的加持，62 亿参数的 ChatGLM-6B 已经能生成相当符合人类偏好的回答。更多信息请参考我们的[博客](https://chatglm.cn/blog)。

不过，由于ChatGLM-6B的规模较小，目前已知其具有相当多的[**局限性**](#局限性)，如事实性/数学逻辑错误，可能生成有害/有偏见内容，较弱的上下文能力，自我认知混乱，以及对英文指示生成与中文指示完全矛盾的内容。请大家在使用前了解这些问题，以免产生误解。

源仓库地址 [https://github.com/THUDM/ChatGLM-6B](https://github.com/THUDM/ChatGLM-6B)

```python
from transformers import AutoModel, AutoTokenizer  # 导入transformers库中的自动模型和分词器加载工具
# import gradio as gr  # 导入gradio库用于web界面（此处未用，默认注释）

# ========================
# 加载ChatGLM-6B模型和分词器
# ========================

# 下面两行用于直接从huggingface官网在线下载模型权重和分词器
# 如果是首次运行，会自动从"THUDM/chatglm-6b"项目下载权重和配置，并保存在缓存目录（~/.cache/huggingface/hub/...）
tokenizer = AutoTokenizer.from_pretrained("THUDM/chatglm-6b", trust_remote_code=True)  
# AutoTokenizer会自动加载适配该模型的tokenizer，trust_remote_code=True表示允许加载官方自定义代码

model = AutoModel.from_pretrained("THUDM/chatglm-6b", trust_remote_code=True) \
    .quantize(4) \            # 使用4-bit量化模型参数，降低显存占用，适合在消费级GPU运行
    .half() \                 # 将模型权重类型转为半精度float16，加速推理、节省显存（NVIDIA显卡支持）
    .cuda()                   # 将模型加载到GPU（cuda设备）上运行以提升速度

# 如果想从本地加载已经下载好的权重，则将上面两行注释，取消下面两行注释，并保证chatglm-6b文件夹和notebook在同一目录下
# tokenizer = AutoTokenizer.from_pretrained("./chatglm-6b", trust_remote_code=True)
# model = AutoModel.from_pretrained("./chatglm-6b", trust_remote_code=True).quantize(4).half().cuda()

# ======================
# 显示模型结构
# ======================

print('-'*50)   # 打印分隔线，便于分隔输出内容
print(model)    # 输出加载的模型结构（层次、参数数量等信息）
print('-'*50)   # 打印分隔线

# 设置模型为推理模式，禁用dropout等训练特性，确保推理一致性和效率
model = model.eval()

# ========================
# 聊天窗口显示参数设定
# ========================

MAX_TURNS = 20        # 一次对话最多支持的轮数（单轮问答为一轮）
MAX_BOXES = MAX_TURNS * 2  # 聊天对话框的总数（每轮包含“用户问题”和“AI回复”各一个文本框，所以是2倍）
```

```python
def predict(input, history=None):
    if history is None:
        history = []
    response, history = model.chat(tokenizer, input, history)
    print(response)
```
