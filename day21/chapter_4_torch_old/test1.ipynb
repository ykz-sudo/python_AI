```python
from collections import namedtuple

# 创建一个 namedtuple 类型
Car = namedtuple('Car', ['make', 'model', 'year'])

# 创建 Car 对象
my_car = Car('Toyota', 'Camry', 2020)

# 通过字段名访问元素
print(my_car.make)  # 输出：Toyota
print(my_car.model)  # 输出：Camry
print(my_car.year)  # 输出：2020

# 通过索引访问元素
print(my_car[0])  # 输出：Toyota
print(my_car[1])  # 输出：Camry
print(my_car[2])  # 输出：2020

# 直接打印对象
print(my_car)  # 输出：Car(make='Toyota', model='Camry', year=2020)
```
