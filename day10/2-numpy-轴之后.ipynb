## 6.2.1 求和

```python
import numpy as np
arr25 = np.arange(1, 13).reshape(3, 4)
print(arr25)
# 求所有数据的和
print(arr25.sum())

# 按轴求和
# axis = 0, 表示按列求和
print(arr25.sum(axis=0))

# axis=1, 表示按行求和
print(arr25.sum(axis=1))
```

## 7.1 数组的添加

```python
arr27 = np.array([[1,2,3],[4,5,6]])
print(arr27)

# 数组的添加
# 1. 默认情况下，展开为一维数组再添加
arr28 = np.append(arr27, np.array([7,8,9]))
print(arr28)

# 2. 指定轴添加
# axis = 0, 沿着0轴添加元素
arr29 = np.append(arr27, np.array([[7,8,9]]), axis=0)
print(arr29)

# axis=1, 沿着1轴添加元素
arr30 = np.append(arr27, np.array([[7],[8]]), axis=1)
print(arr30)
```

```python
arr28 = np.array([[1,2,3],[4,5,6]])
print(arr28)

# 数组的插入
# 1. 默认情况下，展开为一维数组再插入
# 2. 插入的元素必须和数组的元素形状相同

# 2表示位置2
# 7表示插入7
arr29 = np.insert(arr28, 2, 7)
print(arr29)

# 指定轴插入
# axis = 0, 沿着0轴插入元素
# 索引0，表示插入第一行
arr30 = np.insert(arr28, 0, np.array([[7,8,9]]), axis=0)
print(arr30)
```

## 7.2 数组的删除

```python
arr28 = np.array([[1,2,3],[4,5,6]])
print(arr28)

# 1. 默认情况下，展开为一维数组再删除, 2表示删除索引为2的元素
arr31 = np.delete(arr28, 2)
print(arr31)
print('-'*20)

# 2. 指定轴删除
# axis = 0, 沿着0轴删除元素
# 0表示删除索引为0的元素, 也就是删除第一行
arr32 = np.delete(arr28, 1, axis=0)
print(arr32)
```

## 7.3 数组的去重

```python
arr29 = np.array([1,2,3,4,5,5,5,6,7,8,9,9]).reshape(3,4)
print(arr29)

# 1. 直接去重
# 默认情况下，展开一维数组再去重，得到的也是一维数组
arr30 = np.unique(arr29)
print(arr30)

# 2. 按轴去重
# axis = 0, 沿着0轴去重
# 添加一行重复数据
arr31 = np.append(arr29, np.array([[1,2,3,4]]), axis=0)
print(arr31)

# 去重，去除重复数据
arr32 = np.unique(arr31, axis=0)
print(arr32)
```

## 7.4 拼接与分割

```python
# 数组的拼接
arr33 = np.array([[1,2,3],[4,5,6]])
arr34 = np.array([[7,8,9],[10,11,12]])

# 默认情况下，沿着轴0进行拼接
arr35 = np.concatenate((arr33, arr34))
print(arr35)
print('-'*20)

# 指定轴了之后，按轴1拼接
print(np.concatenate((arr33, arr34), axis=1))
```

```python
#数组的堆叠
import numpy as np
a = np.array([1, 2, 3])  # shape: (3,)
b = np.array([4, 5, 6])  # shape: (3,)
# c = np.array([7, 8, 9])  # shape: (3,)

# axis = 0，沿着0轴进行堆叠,可以1轴
np.stack([a, b], axis=1)
```

```python
import numpy as np

# 初始化2个3行4列的数组
a = np.ones((3, 4))
b = np.ones((3, 4)) * 2

# 在0轴堆叠
stack0 = np.stack([a, b], axis=0)
print("0轴堆叠后的尺寸:", stack0.shape)

# 在1轴堆叠
stack1 = np.stack([a, b], axis=1)
print("1轴堆叠后的尺寸:", stack1.shape)

# 在2轴堆叠
stack2 = np.stack([a, b], axis=2)
print("2轴堆叠后的尺寸:", stack2.shape)
```

```python
arr36 = np.arange(1, 13).reshape(3, 4)
print(arr36)

# 1. 默认情况下，按轴0进行分割
# 3表示分割成3份
arr37_list = np.split(arr36, 3)
print(arr37_list)
print(type(arr37_list))

# 2. 按轴1进行分割
arr38 = np.split(arr36, 2, axis=1)
print(arr38)
```

## 7.5 转置与轴滚动

```python
arr39 = np.arange(0, 12).reshape(3, 4)
print("原始数组:")
print(arr39)

# 1. 转置第一种方式
arr40 = arr39.T
print("转置第一种方式:")
print(arr40)

# 2. 转置第二种方式
arr41 = np.transpose(arr39)
print("转置第二种方式:")
print(arr41)


# 4. 轴滚动
# 表示把1轴滚动到0轴的位置
arr43 = np.rollaxis(arr39, 1, 0)
print("轴滚动:")
print(arr43)

# 轴滚动第二个案例
arr44 = np.ones((3, 4, 5, 6))
print(arr44.shape)
# 表示把3轴滚动到1轴的位置
arr45 = np.rollaxis(arr44, 2, 0)
print(arr45.shape)
```
